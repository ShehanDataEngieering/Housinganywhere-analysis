```python
import requests
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import numpy as np
import time
```

```python
def get_requirements(soup):
    try:
        requirements = new_soup.find('div',
                                     attrs={'class': 'MuiGrid-root MuiGrid-container css-q6fy6c-highlightContainer'})

        #Inner navigtable String Object
        requirements_value = requirements.text
    except AttributeError:
        requirements_value = ''
    return requirements_value
```

```python
def get_facilites(soup):
    try:
        facilities = new_soup.find('div', attrs={'MuiGrid-root css-1v5feww-container'})

        facilities_value = facilities.text


    except AttributeError:
        facilities_value = ''
    return facilities_value
```

```python
def get_paragraph(soup):
    try:
        paragraph = soup.find('div',
                              attrs={'MuiPaper-root MuiPaper-outlined MuiPaper-rounded MuiCard-root css-11xz2q3-card'})
        paragraph_value = paragraph.text

    except AttributeError:
        paragraph_value = ""
    return paragraph_value
```

```python
# Function to extract Product Title
def get_title(soup):
    try:
        # Outer Tag Object
        title = soup.find("span",
                          attrs={'class': 'MuiTypography-root MuiTypography-heading/desktop/h4-semibold css-1ql5bbl'})

        # Inner Navigatable String Object
        title_value = title.text

        # Title as a string value
        title_string = title_value.strip()

    except AttributeError:
        title_string = ""

    return title_string
```

```python
def get_price(soup):
    try:
        time.sleep(20)
        price_element = soup.find("p", attrs={
            'class': 'MuiTypography-root MuiTypography-h2 css-2fx1u5-h2-overflow-initial-color-mint-price',
            'data-test-locator': 'Listing/Price/Price'
        })
        price_value = price_element.text

        print(price_value)
    except AttributeError:
        price_value = ""

    return price_value
```

```python
def get_rules(soup):
    try:
        rule = soup.find('ul', attrs={'css-lma2a3-list'})
        rule_value = rule.text
    except AttributeError:
        rule_value = ""

    return rule_value
```

```python
def scrape_links(driver):
    # Get the rendered HTML
    html = driver.page_source

    # Parse the HTML with BeautifulSoup
    soup = BeautifulSoup(html, 'html.parser')

    # Find all links with the specified class
    links = soup.find_all('a', attrs={'class': 'css-1a3e697-cardLink'})

    # Print the links or their href attributes
    for link in links:
        print(link.get('href'))


if __name__ == '__main__':
    # Initialize the WebDriver
    driver = webdriver.Chrome()

    # The webpage URL
    URL = "https://housinganywhere.com/s/Gothenburg--Sweden"

    # Open the webpage using Selenium
    driver.get(URL)

    try:
        WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'a.css-1a3e697-cardLink'))
        )
    except Exception as e:
        print(f"Error: {e}")
        driver.quit()
        exit()

    # Scrolling to load more content
    last_height = driver.execute_script("return document.body.scrollHeight")

    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        last_height = new_height

    # Scrape links from the first page
    scrape_links(driver)

    # Pagination handling
    page_number = 2

    try:

        # Find the next page button and click it
        next_button = WebDriverWait(driver, 50).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.MuiPaginationItem-root'))
        )

        next_button.click()

        # Wait for the content to load
        WebDriverWait(driver, 50).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'a.css-1a3e697-cardLink'))
        )

        # Scrape links from the current page
        print(f"Scraping page {page_number}:")
        scrape_links(driver)

        page_number += 1  # Move to the next page
    except Exception as e:
        print("No more pages to scrape.")

```

```python
# Store the links
links_list = []

# Loop for extracting links from Tag Objects
for link in links:
    print(link.text)
    try:
        href = link['href']
        if not href:
            raise ValueError("Found a link with no href attribute")
        links_list.append(href)
    except Exception as e:
        print(f"An error occurred with link extraction: {e}")

d = {"title": [],  "requirements": [], "paragraph": [], "facilities": [], "rules": []}

# Loop for extracting product details from each link 
for link in links_list:
    new_webpage = requests.get('https://housinganywhere.com/' + link)

    new_soup = BeautifulSoup(new_webpage.content, "html.parser")

    # Function calls to display all necessary product information
    d['title'].append(get_title(new_soup))
    #d['price'].append(get_price(new_soup))
    d['requirements'].append(get_requirements(new_soup))
    d['paragraph'].append(get_paragraph(new_soup))
    d['facilities'].append(get_facilites(new_soup))
    d['rules'].append(get_rules(new_soup))

```

```python
#remove facilities in section
if 'facilities' in d:
    d['facilities'] = [item.replace('Facilities', '').strip() for item in d['facilities']]
```

```python
# Create an empty list to store the modified strings
modified_facilities = []

# Iterate through each item in the 'facilities' list
for item in d['facilities']:
    # Remove 'Amenities' from each item and add it to the modified list
    modified_facilities.append(item.replace("Amenities", "").replace("\n", ""))

# Update the 'facilities' list in the dictionary with the modified list
d['facilities'] = modified_facilities
```

```python
amazon_df = pd.DataFrame.from_dict(d)
amazon_df.to_csv("Housing.csv", header=True, index=False)
```
